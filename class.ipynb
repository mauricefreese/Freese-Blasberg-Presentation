As said before we can highlight minimum and maximum values if needed

```python
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: red' if v else '' for v in is_max]
```

```python
df.style.apply(highlight_max)
```

```python
# Here were are creating a table to show negative numbers and nan values
np.random.seed(24)
dfn = pd.DataFrame({'A': np.linspace(1, 10, 10)})
dfn = pd.concat([dfn, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)
dfn.iloc[3, 3] = np.nan
dfn.iloc[0, 2] = np.nan

dfn.style
```

So now we are going to manipulate the negative and positive numbers

```python
def color_negative(n):
    '''
    This takes a vlue and it will return 
    a str with CSS property, in this case 'color:red'
    for negative and 'color:green' for positive
    '''
    color = 'red' if n < 0 else 'green'
    return 'color: %s' % color
```

To call the Python function to enable it to establish our desired CSS properties we are goin got use the `Styler` objects attribute, `.applymap`. 

```python
dfn.style.applymap(color_negative)
# if we were not dealing with something like stock quotes, we can make the positive number black.
```

We can also use the `.format` attribute to utilize f-strings to display data

```python
dfn.style.format({'A':'{:.2%}', 'C' :'{:.3f}', 'E':'{:%}'}) # here we restrict values to 2 places and 3 after the decimal and add a percent sign on different columns
```

Also, we can get rid of the 'nan' and replace with something else or a symbol

```python
dfn.style.format(na_rep=' ') # you can words, characters or symbols
```

### Practice 3
In this I am goint to need you to add the following the the table below:
- Blue for negative numbers
- Replace "nan" values with a zero
- Highlight the minimum color gold

```python
def blue_negative(n):
    '''
    This takes a vlue and it will return 
    a str with CSS property, in this case 'color:red'
    for negative and 'color:green' for positive
    '''
    color = 'blue' if n < 0 else 'black'
    return 'color: %s' % color

#######################################################################
def highlight_min(n):
    '''
    highlight the maximum in a Series gold.
    '''
    is_min = n == n.min()
    return ['background-color: gold' if v else '' for v in is_min]

dfn.style.applymap(blue_negative).apply(highlight_min).format(na_rep='0')
```

As we see we cannot see the black numbers that well because of dark gray background, so lets change the background color.

```python
'''
By itself
dfn.syle.set_properties(***{'background-color': 'white',
                            'border-color: 'gray',
                            'color': 'purple' changes the characters numbers
                            }
'''
dfn.style.applymap(blue_negative).apply(highlight_min).format(na_rep='0').set_properties(**{'background-color': 'white'})
```
